```python
#convert a string that looks like a list into an actual list is using abstract syntax trees.
import ast

# literal_eval. It "evaluates" the string and recognizes the Python data structure inside.
type(ast.literal_eval(df['job_skills'][1]))
```

```python
#pd.notna -This line checks if the cell is "Not Available" (empty).
def clean_list(skill_list):
    if pd.notna(skill_list):
        return ast.literal_eval(skill_list)
    return skill_list #This ensures NaN stays NaN instead of becoming None

```

```python
df['job_skills']=df['job_skills'].apply(clean_list)
```
