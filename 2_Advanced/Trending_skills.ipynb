```python
#importing Libraries 
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df =dataset['train'].to_pandas()
```

```python
#Data Cleanup
df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
```

```python
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x)else x)
```

```python
#creates a new dataframe with filtered column 
df_DA = df[df['job_title_short']=='Data Analyst'].copy()
```

```python
#creates a new column in dataframe 
df_DA['job_posted_month_no'] = df_DA['job_posted_date'].dt.month
```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_explode
```

```python
#Use pivot when you want to take values from a column and turn them into headers (columns). This "widens" your data.
df_DA_pivot =df_DA_explode.pivot_table(index = 'job_posted_month_no',columns='job_skills',aggfunc='size',fill_value=0)
```

```python


df_DA_pivot.reset_index()
```

```python
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot = df_DA_pivot.drop(columns='job_posted_month_no')   
```

```python
df_DA_pivot.iloc[:,:5].plot(kind = 'line')

plt.title('Top 5 Skills for Data Analysts per Month')
plt.xlabel('')
plt.ylabel('Count')
plt.show()
```

```python
#loc creates a new row named total .This calculates the sum of every column in your DataFrame.
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
```

```python
#The .loc indexer is used to access a row by its label.
df_DA_pivot=df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending = False).index]
```

```python
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.drop('Total')
```

```python
df_DA_pivot
```

```python
#You are grabbing every row but only the first five columns of your pivot table.
df_DA_pivot.iloc[:,:5].plot(kind = 'line')

plt.title('Top 5 Skills for Data Analysts per Month')
plt.xlabel('')
plt.ylabel('Count')
plt.show()
```
