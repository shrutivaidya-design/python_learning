```python
#importing Libraries 
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df =dataset['train'].to_pandas()
```

```python
#Data Cleanup
df['job_posted_date']= pd.to_datetime(df['job_posted_date'])
df.info()
```

```python
#Accessing rows and columns 
df.loc[:10,'salary_rate':'salary_hour_avg']
```

```python
median_salary_year =df.salary_year_avg.median()
median_salary_hour = df.salary_hour_avg.median()
```

```python
df_filled = df
```

```python
df_filled['salary_year_avg']=df_filled.salary_year_avg.fillna(median_salary_year)
```

```python
df_filled['salary_hour_avg']=df.salary_hour_avg.fillna(median_salary_hour)
```

```python
df_filled.loc[:10,'salary_rate':'salary_hour_avg']
```

```python
df_unique = df_filled
```

```python
df_unique =df_unique.drop_duplicates(subset =['job_title','company_name'] )
print('Length of original df:',len(df_filled))
print('Length of drop duplicates df:',len(df_unique))
print('Rows droped:',len(df_filled)-len(df_unique))
```
